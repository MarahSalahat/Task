```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('imdb_top_1000.csv')

df.describe
```

```python
edited_df = df.drop(columns=['Poster_Link', 'Released_Year', 'Certificate', 'Runtime', 'Overview', 'Meta_score', 'Star1', 
                      'Star2', 'Star3', 'Star4', 'Gross'])
edited_df.fillna(edited_df.mean())
```

```python
genre_counts = df['Genre'].value_counts()


genre_counts.idxmax()
```

```python
count_df=pd.DataFrame({'Genre': genre_counts.index, 'Count': genre_counts.values})
print(count_df)
```

```python
count_df.plot.bar(x='Genre', y='Count', figsize=(30,12),legend=False)

plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Count of Movies/TV Shows by Genre')
plt.xticks(rotation=90)

plt.show()
```
